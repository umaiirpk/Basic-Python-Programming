## Input Function
If we want data to be given by user, Input fucntion is used. syntax: input("write a statement which guides user what to type").
Note: By default the argument of input function is string data type but it can be converted into other data type using typecasting.

```python
x = input("Enter some random integer number")
y = input("Enter some random Decimal number")
z = input("Enter your name")
print(x , type(x))
print(y,type(y))
print(z,type(z))
```

```python
# Typecasting data type of input function
a = int(x)
b = float(y)
print(a,type(a))
print(b,type(b))
```

## Print Formatting

```python
# Example:
x = int(input("Please type your age"))
y = x + 25
print("Ali is",x,"years old and his father is",y,"years old.")
```

```python
print("Ali is {a} years old and his father is {b} years old.".format(a=x,b=y))
```

```python
print("Ali is {0} years old and his father is {1} years old.".format(x,y))
```

```python
print("Ali is {} years old and his father is {} years old.".format(x,y))
```
