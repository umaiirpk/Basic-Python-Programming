# Data Types
Variable can store following build-in data types in python:
1. Text type: String (str)
2. Numeric types: integer (int), float, complex
3. Sequence types: list, tuple, range
4. Mapping type: Dictionary (dict)
5. Set types: set, frozenset
6. Boolean type: bool
7. Binary types: bytes, bytearray, memoryview      
                   Note: Type Function i.e. type() can be use to indetify the data type of a variable

```python
# string data type
name = "muhammad Umair"
sub = ' physics '
print(name)
print("Data type of variable 'name' is", type(name))
print("Data type of variable 'sub' is", type(sub))
```

```python
# Integer Date type
rollno = 207811
print(type(rollno))
```

```python
# Float Data Type
cgpa = 3.73
print(type(cgpa))
```

```python
# Complex Data Type
numb = 3+4j
print(type(numb))
```

```python
# List Data Type
x = [ 1, 2, 3, " number " , 4 ]
type(x)
```

```python
# Tuple Data Type
y = ( 1, 2, 3, " number " , 4 )
type(y)
```

```python
s = { 1, 3.3, " hello" }
type(s)
```

```python
d = { "name": "Muhammad Umair", "roll no":207811, "cgpa":3.37 }
type(d)
```

### Type Casting
Type casting is the interconvercion of "int", float and string datatype into one another.

```python
t = 12
type(t)
c = str(t)
print(c)
type(c)
```

```python
b = 3.73
print(type(b))
new = int(b)
print(type(new))
print(new)
```

```python
n = 4
print(type(n))
f = float(n)
print(f)
print(type(f))
```

```python
name = "umair"
print(name, type(name))
no = int(name)
print(no, type(no)
```

```python
cgpa = 3.73
print(cgpa,type(cgpa))
rnd = str(cgpa)
print(type(rnt))
```

```python
# Note: Following typecasting is not valid
# string to float
# float to string
# string to int
```
