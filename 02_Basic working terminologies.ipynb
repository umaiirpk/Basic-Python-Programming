## Comments in python

#### Comments are the lines written inside python script which are not executed.
Comments are very usefull for writting a well structured python script.
For single line comment in python we use hash(#) symbol at start of line or at end of line where we want to comment.
Ctrl + / is shortkey for single line comment

```python
print("This is my first python program")
# print("My name is Muhammad Umair")
```

### Multiline comment in python
If you want to comment multiple lines, just select all lines and the press the shortkey Ctrl + /. Do the same to uncomment
a single line or multiple lines.

```python
print("This is my first python program.")
# print("My name is Muhammad Umair.")
# print(" I teach Physics.")
# print(" I live in Abbottabad.")
```
